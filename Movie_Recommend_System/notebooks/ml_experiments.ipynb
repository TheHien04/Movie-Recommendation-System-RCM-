# ML Experiments — Cinemate Hybrid Recommender

Reproduce CBF, CF, and hybrid ranking.

```python
import sys
from pathlib import Path

backend = Path.cwd().parent / "backend"
if not backend.exists():
    backend = Path.cwd() / "backend"
sys.path.insert(0, str(backend))

from app.ml.content_based import ContentBasedRecommender
from app.ml.collaborative import CollaborativeRecommender
from app.ml.hybrid import HybridRecommender

query = "science fiction space adventure"
cbf = ContentBasedRecommender()
cf = CollaborativeRecommender()
hybrid = HybridRecommender()

print("CBF:", list(cbf.recommend_from_query(query, 5).items()))
print("CF:", list(cf.recommend_from_genres(("science fiction",), 5).items()))
print("Hybrid:", hybrid.recommend(query, {"genres": ["science fiction"]}, top_n=5))
```
